```python
def fV(df, region, categoria, monto_minimo):
    filtro = df[
        (df['region'] == region) &
        (df['categoria'] == categoria) &
        (df['monto_bruto'] >= monto_minimo)
    ]
    return filtro
```

```python
r1 = fV(df, 'Norte', 'Tecnología', 5000)
print(r1)

r2 = fV(df, 'Sur', 'Hogar', 10000)
print(r2)

r3 = fV(df, 'Norte', 'Alimentos', 2000)
print(r3)
```

```python
resumen = df.groupby('region').agg(
    suma_total=('monto_bruto', 'sum'),
    promedio=('monto_bruto', 'mean'),
    maximo=('monto_bruto', 'max'),
    minimo=('monto_bruto', 'min'),
    cantidad_transacciones=('monto_bruto', 'count')
).round(2)

resumen
```

```python
resumen2 = resumen.reset_index().sort_values(
    by='suma_total', 
    ascending=False
)

resumen2
```

```python
resumen_region_mes = df.groupby(['region', 'mes']).agg(
    suma_monto=('monto_bruto', 'sum'),
    suma_cantidad=('cantidad', 'sum')
).reset_index()

resumen_region_mes.head(10)
```
