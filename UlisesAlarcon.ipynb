```python
class Empleado:
    def __init__(self, nombre, empresa, salario_base):
        self.nombre = nombre
        self.empresa = empresa
        self.salario_base = salario_base
    def describir(self):
        return f"{self.nombre} - Salario: {self.salario_base}"
    def calcular_salario(self):
        return self.salario_base
class GerenteVentas(Empleado):
    def __init__(self, nombre, empresa, salario_base, bono_ventas):
        super().__init__(nombre, empresa, salario_base)
        self.bono_ventas = bono_ventas
    def calcular_salario(self):
        return self.salario_base + self.bono_ventas
class ObreroPorHora(Empleado):
    def __init__(self, nombre, empresa, horas_trabajadas, valor_hora):
        salario_base = horas_trabajadas * valor_hora
        super().__init__(nombre, empresa, salario_base)
        self.horas_trabajadas = horas_trabajadas
        self.valor_hora = valor_hora
    def calcular_salario(self):
        return self.salario_base    
```

```python
emp1 = Empleado("ulises", "entel", 120000)
emp2 = GerenteVentas("María", "Entel", 900000, 150000)
emp3 = ObreroPorHora("Carlos", "Sodimac", 160, 5000)
```

```python
print (emp1.describir())
print (emp2.describir())
print (emp3.describir())
```

```python
class Electrico:
    def __init__(self, bateria_kwh):
        self.bateria_kwh = bateria_kwh
    def ficha_tecnica(self):
        return f"Batería: {self.bateria_kwh} kWh"
class Combustion:
    def __init__(self, cilindrada_cc):
        self.cilindrada_cc = cilindrada_cc
    def ficha_tecnica(self):
        return f"Cilindrada: {self.cilindrada_cc} cc"
class Vehiculo:
    def __init__(self, marca, modelo, año):
        self.marca = marca
        self.modelo = modelo
        self.año = año
    def ficha_tecnica(self):
        return f"{self.marca} {self.modelo} ({self.año})"
class Auto(Vehiculo):
    def ficha_tecnica(self):
        return "Auto - " + super().ficha_tecnica()
class Moto(Vehiculo):
    def ficha_tecnica(self):
        return "Moto - " + super().ficha_tecnica()
class AutoElectrico(Auto, Electrico):
    def __init__(self, marca, modelo, año, bateria_kwh):
        Auto.__init__(self, marca, modelo, año)
        Electrico.__init__(self, bateria_kwh)
    def ficha_tecnica(self):
        return super().ficha_tecnica() + " | " + Electrico.ficha_tecnica(self)
```

```python
auto = Auto("vosvagen", "escarabajo", 1980)
moto = Moto("kawasaki", "3", 2021)
auto_elec = AutoElectrico("autito electrico", "2", 2022, 75)

print(auto.ficha_tecnica())
print(moto.ficha_tecnica())
print(auto_elec.ficha_tecnica())

print("MRO de AutoElectrico:")
print(AutoElectrico.__mro__)
```

```python
import math

class Figura(ABC):
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimetro(self):
        pass
class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio
    def area(self):
        return math.pi * self.radio ** 2
    def perimetro(self):
        return 2 * math.pi * self.radio
class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
    def area(self):
        return self.base * self.altura
    def perimetro(self):
        return 2 * (self.base + self.altura)
class Triangulo(Figura):
    def __init__(self, base, altura, hipotenusa):
        self.base = base
        self.altura = altura
        self.hipotenusa = hipotenusa
    def area(self):
        return (self.base * self.altura) / 2
    def perimetro(self):
        return self.base + self.altura + self.hipotenusa
def calcular_total(figuras: list):
    total = 0
    for figura in figuras:
        total += figura.area()
    return total
```

```python
f1 = Circulo(5)
f2 = Rectangulo(4, 6)
f3 = Triangulo(3, 4, 5)

figuras = [f1, f2, f3]

print("Área total:", calcular_total(figuras))
```

```python

class MetodoPago(ABC):
    @abstractmethod
    def procesar(self, monto):
        pass
    @abstractmethod
    def reembolsar(self, monto):
        pass
class PagoTarjeta(MetodoPago):
    def __init__(self, numero_tarjeta, titular):
        self.numero_tarjeta = numero_tarjeta
        self.titular = titular
    def procesar(self, monto):
        print(f"Procesando pago de ${monto} con tarjeta de {self.titular}")
    def reembolsar(self, monto):
        print(f"Reembolsando ${monto} a la tarjeta de {self.titular}")
class PagoTransferencia(MetodoPago):
    def __init__(self, cuenta_origen, cuenta_destino):
        self.cuenta_origen = cuenta_origen
        self.cuenta_destino = cuenta_destino
    def procesar(self, monto):
        print(f"Transfiriendo ${monto} desde {self.cuenta_origen} a {self.cuenta_destino}")
    def reembolsar(self, monto):
        print(f"Revirtiendo transferencia de ${monto} desde {self.cuenta_destino} a {self.cuenta_origen}")
class PagoEfectivo(MetodoPago):
    def procesar(self, monto):
        print(f"Pago en efectivo de ${monto} recibido")
    def reembolsar(self, monto):
        print(f"Devolviendo ${monto} en efectivo")
def ejecutar_pago(metodo: MetodoPago, monto: float):
    metodo.procesar(monto)



```

```python
pago1 = PagoTarjeta("123", "elpepe")
pago2 = PagoTransferencia("Cuenta1", "Cuenta2")
pago3 = PagoEfectivo()

ejecutar_pago(pago1, 50000)
ejecutar_pago(pago2, 75000)
ejecutar_pago(pago3, 10000)
```
