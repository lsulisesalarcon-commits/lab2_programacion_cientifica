```python
from functools import reduce
import random
```

```python
# Actividad 1 – Uso de lambda

lista=[2,4,6]
ecu = list(map(lambda x:x**2 + 2*x + x, lista))
```

```python
print (ecu)
```

```python
# Actividad 2 – Transformación con map()

temperaturas = [15, 18, 21, 35, 42, 12, 9, 30, 27]
```

```python
kelvin = list(map(lambda x: x + 273.15, temperaturas))
```

```python
print (kelvin)
```

```python
velocidades = [10, 20, 30, 40, 50]

Metros = list(map(lambda x: x * 1000, velocidades))
```

```python
print (Metros)
```

```python
# Actividad 3 – Filtrado con filter()

Temperaturas = [15, 18, 21, 35, 42, -10, 9, 30, 150]
```

```python
datos_validos = list(filter(lambda x: -20 <= x <= 120, temperaturas))
```

```python
print (datos_validos)
```

```python
valores = [2,5,8,11,14,17,20]
```

```python
par = list(filter(lambda x: x % 2 == 0, valores))
```

```python
print (par)
```

```python
# Actividad 4 – Agregación con reduce()

numeros = [5,10,15,20]
```

```python
resultado = reduce(lambda x,y: x+y, numeros)
```

```python
print (resultado)
```

```python
val = [2,3,4,5]
```

```python
resul = reduce(lambda x,y: x*y, val)
```

```python
print (resul)
```

```python
# Actividad 5 – Pipeline de Procesamiento

data = [12,15,22,35,41,8,3,55,62,90]
```

```python
result = reduce(lambda x,y: x+y,
        map(lambda x: x*2,
            filter(lambda x: x>=10, data)))
                
```

```python
print (result)
```

```python
# Actividad 6 – Simulación de Big Data

datos = [random.randint(0,100) for _ in range(1000000)]
```

```python
Ergebnis = reduce(lambda x,y: (x+y)/len(datos),
        map(lambda x: x**2,
            filter(lambda x: x>50, datos)))
```

```python
print (Ergebnis)
```

```python
# Desafío Avanzado

sensores = [random.uniform(10,80) for _ in range(500000)]
```

```python
kekka = list(map(lambda x: x + 273.15,
            filter(lambda x: x>=10 and x<=70, sensores)))
```

```python
prom = reduce(lambda x, y: x + y, kekka)/len(kekka)
```

```python
print("{:.2f}".format(prom))
```
